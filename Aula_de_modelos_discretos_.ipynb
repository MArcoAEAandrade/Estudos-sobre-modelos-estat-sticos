Exercício 1

Uma caixa contém 30 peças, sendo 6 defeituosas.
Retiram-se 5 peças sem reposição.

Pergunta: qual a probabilidade de sair no máximo 1 defeituosa?

$$
P(X = k)=
\frac{\binom{K}{k}\binom{N-K}{n-k}}
{\binom{N}{n}}
$$

```python
#dados
#P(X <=1) = P(X = 0) + P(X=1)
# "X = n° de peças defeituosas"


N = 30 # peças no geral
K = 6 #peças defeituosas
n = 5 # retirar sem reposição

k= 0
px0 = math.comb(K,k) * math.comb(N-K,n-k)/math.comb(N,n)

k = 1
px1 = math.comb(K,k) * math.comb(N-K,n-k)/math.comb(N,n)

px0 + px1
round (px0 + px1,5)
```

O lote é aceito se houver no máximo 1 defeituosa na amostra.

Pergunta: qual a probabilidade de o lote ser aceito?

```python
#dados
N = 40 #peças
K = 5 #peças defeituosas
n = 6 #retira-se peças na inspeção

#P(X <= 1) = P(X = 0) + P(X = 1)

k = 0
px0 = math.comb(K,k) * math.comb(N-K,n-k)/math.comb(N,n)

k = 1
px1 = math.comb(K,k) * math.comb(N-K,n-k)/math.comb(N,n)

px0 + px1
round (px0 + px1,5)
```
