# Distribuição Exponencial

Função densidade de probabilidade (PDF)

$$f(x)=\lambda e^{-\lambda x}, \quad x \geq 0$$

---
Falha acumulada

$$P(X \leq x)=1-e^{-\lambda x}$$

---
Sobrevivência

$$P(X > x)=e^{-\lambda x}$$

---
Média

$$\mu=\frac{1}{\lambda}$$

---

Variância
$$\sigma^2=\frac{1}{\lambda^2}$$

```python
import math
from scipy.integrate import quad

# dados
u = 100
lambd = 1/u

# função densidade exponencial
f = lambda x: lambd * math.exp(-lambd * x)

# integral de 0 até 50
p_menor_50, erro = quad(f, 0, 50)

# complemento
p_maior_50 = 1 - p_menor_50

print("P(X <= 50):", p_menor_50)
print("P(X > 50):", p_maior_50)
```

$$P(X =k) = \frac{n!}{(n-k)!k!} p^k (1-p)^ (n-k)$$


```python
#b) P(X =  3) "X = Latas rejeitadas"
import math
n = 20
k = 3
u = 990 #gramas
dp = 10 #gramas
a = 976 #gramas
p = st.norm.cdf(a, loc=u, scale=dp)

resultado = round(math.factorial(n)/(math.factorial(k)*math.factorial(n-k)) * p**k * (1-p)**(n-k) ,5)
print('a PROBABILIDADE  da 3 lata serem rejeitada em 20 observadas é ', resultado)
```

```python
# dados (forma mais ágil)
u = 990 #gramas
dp = 10 #gramas
a = 976 #gramas

# binomial
n = 20
k = 3

# probabilidade de rejeição
p = st.norm.cdf(a, loc=u, scale=dp)

# probabilidade binomial
resultado = round(st.binom.pmf(k, n, p),6)
print('a PROBABILIDADE  da 3 lata serem rejeitada em 20 observadas é ', resultado)
```

```python
#dados
#a) P(4 <= X < = 5)
import numpy as np
import scipy.integrate
u = 4 # minutos
lambd = 1/u
f = lambda x: lambd * np.exp(-lambd * x)
p, erro = scipy.integrate.quad(f, 4, 5)
print(round((p),8))
```

```python
import scipy.stats as st

u = 4

# P(4 <= X <= 5)
p = st.expon.cdf(5, scale=u) - st.expon.cdf(4, scale=u)
print(round((p),8))
```

```python
#b) P(X > = 5) = 1 - P(X < 5)
import numpy as np
import scipy.integrate
u = 4 # minutos
lambd = 1/u
f = lambda x: lambd * np.exp(-lambd * x)
p, erro = scipy.integrate.quad(f, 0, 5)
print(round((1-p),8))
```

```python
#b) P(X > = 6)
import scipy.stats as st

u = 4

p = st.expon.sf(6, scale=u)  # survival function = P(X > 5)
print(round(p, 8))
```

```python
#b) P(X > = 6) = 1 - P(X < 6)
import scipy.stats as st

u = 4

p = 1 - st.expon.cdf(6, scale=u)
print(round(p, 8))
```
