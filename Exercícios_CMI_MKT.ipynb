Em uma campanha de e-mail marketing, sabe-se que cada cliente tem probabilidade p=0,12 de clicar no anúncio, independentemente dos demais acessos.

Responda:

(a) Considere o envio para n=15 clientes. Qual a probabilidade de exatamente k=3 clientes clicarem no anúncio?

(b) Qual a probabilidade de o primeiro clique ocorrer exatamente no 5º envio?

(c) Qual a probabilidade de o 4º clique ocorrer exatamente no 10º envio?

(d) Um banco de dados possui N=80 leads, dos quais K=18 já interagiram anteriormente com a marca. Seleciona-se uma amostra de n=6 leads sem reposição. Qual a probabilidade de exatamente k=2 terem interagido antes?

```python
#dados
#a) (PX = 3 )
p = 0.12 #clicar em um anúncio
n = 15
k =3 # clientes cliquem no anúncio

resultado = round(st.binom.pmf(k,n,p),5)
print(resultado)


```

```python
#dados
#b) P(X = 1° cliente esteja na 5° envio )
p = 0.12 #clicar em um anúncio
k =5 # clientes cliquem no anúncio

resultado = st.geom.pmf(k,p)
print(resultado)

```

```python
import scipy.stats as st
#c) (PX = 4° clicle esteja na 10° envio )
p = 0.12
r = 4  # clientes cliquem no anúncio
x = 6   # fracassos antes do 4º sucesso

resultado = st.nbinom.pmf(x, r, p)

print(round(resultado, 6))
```

```python
import scipy.stats as st
#d) (PX = 4° clicle esteja na 10° envio )
p = 0.12
N = 80 # leads (população)
K = 18  # já interagiram (sucessos da população)
n = 6 # amostra
k = 2  # clientes que interagiram antes (sucuessos da amostra)


resultado = st.hypergeom.pmf(k, N, K, n)

print(round(resultado, 6))
```

Em um aplicativo de delivery, o número médio de pedidos recebidos é de λ=5 pedidos por minuto.

Responda:

(a) Qual a probabilidade de ocorrerem exatamente k=7 pedidos em um minuto?

(b) Qual a probabilidade de ocorrerem no máximo 3 pedidos em um minuto?

(c) Qual a probabilidade de ocorrerem pelo menos 12 pedidos em dois minutos?

(d) Qual o número esperado de pedidos em 8 minutos? E o desvio padrão?

```python
#a) P(X = 7)
lambd = 5
k = 7

resultado = st.poisson.pmf(k,lambd)
print(resultado)
```

```python
#b) P(X <= 3)
lambd = 5
k = 3

resultado = st.poisson.cdf(k,lambd)
print(resultado)
```

```python
#c) P(X >= 12)
lambd = 5
k = 11

resultado = st.poisson.sf(k,lambd)
print(resultado)
```

```python
from math import sqrt
#d) esperança e desvio padrão
E =  lambd #esperança
DP = sqrt(E) #desvio padrão

print('A espera é: ',E)
print('O desvio padrão é', DP)
```

O tempo que um usuário permanece navegando em um e-commerce segue aproximadamente uma distribuição com média μ=18 minutos e desvio padrão σ=4 minutos.

Responda:

(a) Qual a probabilidade de um usuário permanecer menos de 15 minutos no site?

(b) Qual a probabilidade de o usuário permanecer entre 16 e 22 minutos?

(c) Determine o tempo t tal que 95% dos usuários permaneçam mais do que t minutos no site.

(d) Qual o valor esperado e a variância dessa distribuição?

```python
#dados
#a) P(X < 15 )
u = 18 #minutos
DP = 4 # minutos
a = 15 #minutos

resultado = st.norm.cdf(a, loc=u, scale=DP)
print(resultado)


```

```python
#dados
#b) P( 16 <= X <= 22 )
u = 18 #minutos
DP = 4 # minutos
a = 16 #minutos
b = 22 #minutos

resultado = st.norm.cdf(b, loc=u, scale=DP) - st.norm.cdf(a, loc=u, scale=DP)
print(resultado)
```

```python
#dados
#c) P(T > t ) = 1 - P(T < t ) =  0.95 = 1 - 0.95 = 0.05
u = 18 #minutos
DP = 4 # minutos
a = 0.05 #minutos

resultado = st.norm.ppf(a, loc=u, scale=DP)
print(resultado)
```

```python
#dados
#d) Qual o valor esperado e a variância dessa distribuição?
u = 18 #minutos
E = u
DP = 4 # minutos
V = DP**2

print('A esperança é: ',E)
print('A variância é', V)

```

O tempo até um cliente cancelar a assinatura de um streaming segue uma taxa média de cancelamento λ=0,08 por mês. Responda:

(a) Qual a probabilidade de um cliente permanecer assinante por mais de 12 meses?

(b) Qual a probabilidade de o cancelamento ocorrer antes de 3 meses?

(c) Qual o valor esperado e a variância do tempo até o cancelamento?

```python
#dados
#a) P(X > 12 )
lambd = 0.08 # taxa de cancelamento por mês
x = 12

resultado = round(st.expon.sf(x, scale=1/lambd),5)
print(resultado)
```

```python
#dados
#b) P(X < 3 )
lambd = 0.08 # taxa de cancelamento por mês
k = 3
resultado =  round(st.expon.cdf(k, scale=1/lambd),5)
print(resultado)
```

```python
#dados
# c) Qual o valor esperado e a variância do tempo até o cancelamento?
lambd = 0.08 # taxa de cancelamento por mês
E = 1/lambd
V = (1/lambd)**2

print('A esperança é: ',E)
print('A variância é', V)
```

Em uma loja virtual, sabe-se que 7% dos pedidos apresentam algum problema logístico (atraso, troca incorreta ou avaria), independentemente dos demais. Responda:

(a) Em n=20 pedidos, qual a probabilidade de exatamente k=1 apresentar problema?

(b) Qual a probabilidade de o primeiro pedido com problema ocorrer exatamente no 6º envio?

(c) Qual a probabilidade de o 2º pedido problemático ocorrer exatamente no 9º envio?

(d) Um centro de distribuição possui N=120 pedidos separados, sendo K=15 com inconsistências. Selecionam-se n=10 pedidos sem reposição. Qual a probabilidade de exatamente k=3 terem

```python
#dados
# a)
p = 0.07 # apresentar algum problema logístico
n = 20 # pedidos
k = 1  #apresentar defeito

resultado = round(st.binom.pmf(k, n , p ),5)
print(resultado)
```

```python
# b)
p = 0.07 # apresentar algum problema logístico
k = 6  #apresentar defeito


resultado = round(st.geom.pmf(k, p ),5)
print(resultado)

```

```python
# c)
p = 0.07 # apresentar algum problema logístico
r = 2  #apresentar defeito no  9º envio?
x = 7 # fracassos(sem deferitos) para os sucessos



resultado = round(st.nbinom.pmf(x, r, p),5)
print(resultado)
```

```python
# d)
N = 120 # pedidos (população)
K = 15  # inconsistências (sucessos da população)
n = 10 # amostra
k = 3  #sucesso da amostra

resultado = round(st.hypergeom.pmf(k, N, K, n),5)
print(resultado)
```
