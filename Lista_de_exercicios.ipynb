```python
#dados
#A) P(X = 2 )

#BINOMINAL
p = 0.1 # probabilidade de ser defeituosos
n = 10 #componentes aleatórios
k = 2 #componentes defeituosos

resultado = round(st.binom.pmf(k, n, p),5)
print(resultado)
```

```python
#B) P(X = 4 ) O PRIMEIRO COMPONENTE DEFEITUOSO NA 4° POSIÇÃO
#GEOMÉTRICA
p = 0.1 # probabilidade de ser defeituosos
k = 4 # tentativa desejada

resultado = round(st.geom.pmf(k, p),5)
print(resultado)
```

$$P(X = k) = \binom{k-1}{r-1} p^r (1-p)^{k-r}$$

```python
#dados
#c) P(X = 3° componente defeituoso estar exatamente na 7° inspeção )

#BINOMINAL - negativa
P = 0.1 # probabilidade de ser defeituosos
k = 7 # némero de eventos
r = 3 #número de sucessos desejados (peças defeitusos)

math.comb(k-1,r-1) * P**r * (1-P)**(k-r)
round(math.comb(k-1,r-1) * P**r * (1-P)**(k-r), 5)
```
